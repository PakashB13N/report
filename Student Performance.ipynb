```python
# Manually inputting the descriptive statistics you provided
data_dict = {
    'Attribute': ['AttendanceRate', 'StudyHoursPerWeek', 'PreviousGrade', 'ExtracurricularActivities', 'FinalGrade'],
    'Mean': [71.7, 47.6, 71.1, 1.5, 80.2],
    'Min': [53, 8, 52, 0, 62],
    '25%': [61.5, 19.5, 60.75, 1, 72],
    '50% (Median)': [68, 58.5, 74, 1.5, 82.5],
    '75%': [85, 68.25, 78.25, 2, 88],
    'Max': [95, 79, 90, 3, 92]
}

# Create a DataFrame
df = pd.DataFrame(data_dict)

# Plot a box plot using seaborn
plt.figure(figsize=(10, 6))
sns.boxplot(data=[df['Mean'], df['Min'], df['25%'], df['50% (Median)'], df['75%'], df['Max']])
plt.xticks([0, 1, 2, 3, 4], df['Attribute'], rotation=45)
plt.title('Box Plot of Descriptive Statistics for Student Data')
plt.xlabel('Attributes')
plt.ylabel('Values')
plt.show()
```

```python
qc = df['FinalGrade'].value_counts()
plt.figure(figsize=(12,6))
plt.bar(qc.index, qc, color="orange")
plt.title("Count Plot of FinalGrade")
plt.xlabel("FinalGrade")
plt.ylabel("Count")
plt.grid(axis="y")
plt.show()
```

```python
PreviousGrade_FinalGrade = df.groupby("PreviousGrade")["FinalGrade"].sum()
PreviousGrade_FinalGrade
```

```python
plt.figure(figsize = [18,9])
PreviousGrade_FinalGrade.plot(kind = "bar", color = "brown")
plt.xlabel("PreviousGrade")
plt.ylabel("FinalGrade")
plt.xticks(rotation = 0)
plt.grid(axis = "y")
plt.title("Final Result")
```

```python
#strike rate in ODIS
plt.figure(figsize=(10, 6))
plt.hist(df['AttendanceRate'], bins=20, color='green', edgecolor='black')
plt.title('Attendance of students', fontsize=16)
plt.xlabel('AttendanceRate', fontsize=14)
plt.ylabel('StudentID', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot for AttendanceRate vs FinalGrade
plt.figure(figsize=(10,6))
plt.scatter(df['AttendanceRate'], df['FinalGrade'], color="blue")
plt.title("Attendance Rate vs Final Grade")
plt.xlabel("Attendance Rate")
plt.ylabel("Final Grade")
plt.grid(True)
plt.show()
```

```python
# Scatter plot for StudyHoursPerWeek vs FinalGrade
plt.figure(figsize=(10,6))
plt.scatter(df['StudyHoursPerWeek'], df['FinalGrade'], color="green")
plt.title("Study Hours per Week vs Final Grade")
plt.xlabel("Study Hours per Week")
plt.ylabel("Final Grade")
plt.grid(True)
plt.show()
```

```python
# Pie chart for FinalGrade distribution
qc = df['FinalGrade'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(qc, labels=qc.index, autopct='%1.1f%%', startangle=90, colors=['red', 'lightblue', 'lightgreen', 'orange', 'yellow'])
plt.title("FinalGrade Distribution")
plt.show()
```
